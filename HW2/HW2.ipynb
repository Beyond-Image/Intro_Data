```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('iris_dataset.csv')
print(df)
```

```python
sns.pairplot(df)
```

```python
cor_mat = df.corr()
print(f'Correlation Matrix: \n {cor_mat} \n')
cov_mat = df.cov()
print(f'Covariance Matrix: \n {cov_mat} \n')
```

```python
boxplot = df.boxplot()
```

```python
sns.boxplot(x = df['sepal_width'])
plt.show()
```

```python
df.describe()
```

```python
IQR = (df['sepal_width'].quantile(0.75)) - (df['sepal_width'].quantile(0.25))
upper_lim = df['sepal_width'].quantile(0.75) + 1.5 * IQR
lower_lim = df['sepal_width'].quantile(0.25) - 1.5 * IQR

out_low = (df['sepal_width'] < lower_lim)
out_high = (df['sepal_width'] > upper_lim)
```

```python
df_filtered = df[~(out_low | out_high)]
```

```python
boxplot = df_filtered.boxplot()
```

```python
sns.pairplot(df_filtered)
```

```python
print(f'Correlation of DF without outliers: \n{df_filtered.corr()} \n \n Covariance of DF wihtout ouliers: \n{df_filtered.cov()} \n')
```
